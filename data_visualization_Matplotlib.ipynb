# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,10,15)
x
```

```python
y = x**2
y
```

```python
#Functional method 
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Sample Plot')
```

# Subplot

This is used for creating multiplot on the same canvas

```python
#Argument takes - (1: Number of rows | 2: Number of columns | 1: The plot number we are refering)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

# Object oriented method

```python
#Object oriented method
fig = plt.figure()
#getting blank axes (canvas). add_axes takes an arguement of left, bottom, width and height
axes = fig.add_axes([0.1,0.5,0.8,0.8])
#Ploting the values
axes.plot(x,y)
#set labels
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Plot usign Object')
```

# Lets try to create canvas and see how their position change with respect to their values

```python
#Both created axes will merge as their values are same
fig = plt.figure()
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
axes2 = fig.add_axes([0.1,0.5,0.8,0.8])
```

```python
#The first one is 0.1 from left side where as axes2 is 0.2 from the left
# axes1 : 0.5 from the bottom whereas axes2: 0.2 from the bottom.
#Same width and height
fig = plt.figure()
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
axes2 = fig.add_axes([0.2,0.2,0.8,0.8])
```

```python
#The first one is 0.1 from left side where as axes2 is 0.4 from the left
# axes1 : 0.5 from the bottom whereas axes2: 0.6 from the bottom.
fig = plt.figure()
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
axes2 = fig.add_axes([0.4,0.6,0.4,0.5])
```

```python
fig = plt.figure()
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
axes2 = fig.add_axes([0.2,0.9,0.3,0.2])

axes1.plot(x,y)
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')
axes1.set_title('Outer_plot')




axes2.plot(x,y,'r')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
axes2.set_title('Inner_plot')
```

# Legend

```python
fig = plt.figure(figsize=(10,5))
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
#
axes1.plot(x,x**2,label='x-square')
axes1.plot(x,x**3,label='x-cube')
#loc is used to select location for legend
axes1.legend(loc=0)
```

# setting colours

```python
fig = plt.figure(figsize=(10,5))
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
#
axes1.plot(x,y,color='green')
```

# setting linewidth, transparency & linestyle

```python
fig = plt.figure(figsize=(10,5))
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
#
axes1.plot(x,y,color='green',linewidth=7, alpha=0.5)
```

```python
fig = plt.figure(figsize=(10,5))
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
#
axes1.plot(x,y,color='green',linewidth=7, linestyle='--')
```

# Marker

```python
#To markout each point
fig = plt.figure(figsize=(10,5))
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
#
axes1.plot(x,y,color='green',linewidth=2,marker='o',markersize=10,markerfacecolor='red',markeredgewidth=3,markeredgecolor='yellow')
```

```python
#To markout each point
fig = plt.figure(figsize=(10,5))
#
axes1 = fig.add_axes([0.1,0.5,0.8,0.8])
#
axes1.plot(x,y,color='red',linewidth=2)
```
